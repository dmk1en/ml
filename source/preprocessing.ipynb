# Preprocess

## 1. Import library

```python
import pandas as pd
from nltk.corpus import stopwords
```

## 2. Preprocessing

### 2.1. Remove stopword

```python
# Load the dataset
df = pd.read_csv('../data/emails.csv')
```

```python
stop_words = set(stopwords.words('english'))
for column in df.columns:
    if column in stop_words:
        df.drop(column, axis=1, inplace=True)

# Remove the columns that are not useful
df.to_csv('../data/Preprocessing.csv', index=False)
```

### 2.2. TF_IFD Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

df['combined_text'] = df.drop(['Email No.', 'Prediction'], axis=1).apply(
    lambda row: ' '.join([f"{word} " * int(row[word]) for word in row.index if row[word] > 0]), axis=1
)

print(df['combined_text'])
```

```python
# Create the TF-IDF vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
vectorizer = TfidfVectorizer()
df1 = pd.read_csv('../data/tfidf_output_cleaned.csv',encoding='latin1')
tfidf_matrix = vectorizer.fit_transform(df1['imp_feature'])

# Convert the tfidf_matrix to a DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

tfidf_df["prediction"] = df1["Label"]

# Save the DataFrame to a CSV file
tfidf_df.to_csv('../data/tfidf_matrix.csv', index=False,encoding='latin1')
```
